En este caso se planteó el interés del equipo por **explorar** qué tipo de volcanes hay en el planeta y el riesgo que representan para la población en las diferentes zonas poblacionales.

Puede resultar para algunos grupos de la población mundial que la actividad volcánica solo sea parte de una noticia en medios de comunicaciones y redes sociales, pero para otra parte de la población es una situación de alto riesgo.

##Normalización de nombres de columnas

Como se resaltó anteriormente los nombres de las columnas no siguen la convención adecuada para el ambiente de desarrollo con Python, por lo que es propio normalizarlos. Para ello nos apoyaremos de la biblioteca 're'.



```python
# Creamos un diccionario con llave=antiguo-nombre y valor=nuevo-nombre
import re
diccionario_convencion = {}
for col in [elem for elem in df.columns]:
  elem = re.sub(" ","_",col)
  elem = re.sub('\W','',elem)
  elem = elem.lower()
  diccionario_convencion[col] = elem
```

```python
# Validamos el diccionario creado.
diccionario_convencion
```

Efectuamos el renombramiento.

```python
df = df.rename(columns = diccionario_convencion)
```

Y validamos.

```python
df.columns
```

##Agrupaciones

Procedemos a identificar todos los tipos de volcanes existentes.

```python
df.groupby('primary_volcano_type')['primary_volcano_type'].count().reset_index(name='count')
```

Vemos que en principio hay 32 tipos de volcanes, sin embargo, podemos notar que algunos de ellos son solamente el plural de los otros valores o corresponden a errores tipográficos, por lo que podemos hacer otra limpieza de los datos para obtener los valores únicos que corresponden realmente.

```python
#Creamos la función que limpia los valores de 'primary_volcano_type'
def clean_values(x):
  x = re.sub('\(\w*\)|\W','',x)
  return x
  
#Aplicamos a la columna
df['primary_volcano_type'] = df['primary_volcano_type'].map(lambda x: clean_values(x))
```

Y agrupamos nuevamente.

```python
volcano_types_df = df.groupby('primary_volcano_type')['primary_volcano_type'].count().reset_index(name='count').sort_values('count', ascending=False).reset_index(drop=True)
volcano_types_df['pct'] = (volcano_types_df['count']/volcano_types_df['count'].sum())*100 
volcano_types_df
```

Vemos que en realidad hay 19 tipos de volcanes.
> <font color='crimson'>**Responde al plateamiento: 4. ¿Qué tipos de volcanes existen?**</font>

```python
categorias_porc = volcano_types_df['pct'][:7].sum()
print(f'Porcentaje de categorías: {int(categorias_porc)}%')
```

Notamos que las primeras 7 categorías agrupan a más del 90% de los volcanes

###Uso de Merge

Y ahora, nos proponemos a obtener la altura media por tipo de volcán.

```python
avg_elevation = df.groupby('primary_volcano_type')['elevation_m'].mean().reset_index(name='avg_elevation')
volcano_types_df = pd.merge(volcano_types_df,avg_elevation, left_on='primary_volcano_type', right_on='primary_volcano_type')
volcano_types_df[['primary_volcano_type','avg_elevation']]
```

## Mapa 1: Volcanes del mundo

En este primer mapa ubicaremos todos los volcanes contenidos en el dataset original.
> <font color='crimson'>**Responde al plateamiento: 5. ¿Dónde se ubican?**</font>

Lo primero que realizamos es crear nuestro mapa base.

```python
#!pip install folium

import folium
world_map = folium.Map(location=[15,-20], zoom_start=2.3, tiles= 'Stamen Terrain')
```

Luego, creamos un objeto `MarkerCluster` y lo anexamos al mapa, este objeto nos permitirá agrupar los volcanes cercanos entre sí lo que nos ayudará a ubicarlos por regiones.

```python
from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster()
world_map.add_child(marker_cluster)
```

Por último, agregamos un marcador para cada volcán, lo agrupamos en el `MarkerCluster` y automáticamente se agrega al mapa.
Y,... ¡vualá!, tenemos nuestro primer mapa.

```python
for lat,lng,name,v_type in zip(df.latitude, df.longitude,df.volcano_name,df.primary_volcano_type):
    marker= folium.features.Marker(
            location= [lat,lng],
            tooltip=name,
            popup=v_type,
            icon= folium.features.CustomIcon(
                icon_image='https://volcano.si.edu/includes/images/volcano_icon_24.png',
                    icon_size=(20,20))
            )
    marker_cluster.add_child(marker)
world_map
```

## Mapa 2: Volcanes de México

En este mapa, nos proponemos a visualizar los volcanes que se encuentran exclusivamente en territorio nacional, y sus zonas de riesgo alto (10 km) y medio (30 km).

La construcción del mapa es similar al anterior, pero tiene sus sutiles diferencias.

Primero, a partir de dataset principal creamos el subconjunto con el que se trabajó.

```python
volcanes_de_mexico = df[df.country=='Mexico']
volcanes_de_mexico
```

Creamos el mapa base y el objeto que agrupará todos los marcadores.

```python
mexico_map = folium.Map(location=[22,-100],zoom_start=5.4, tiles='Stamen Terrain')
ubicaciones = folium.map.FeatureGroup()
```

Ahora, agregamos un identificador para cada volcán y dos círculos representando las zonas de riesgo alto (rojo) y medio (naranja).

```python
# Iteramos dentro del dataset
for lat,lng,name,v_type in zip(volcanes_de_mexico.latitude, volcanes_de_mexico.longitude,volcanes_de_mexico.volcano_name,volcanes_de_mexico.primary_volcano_type):
    ubicaciones.add_child(           # Primer objeto: identificador del volcán
        folium.features.Marker(              
            location= [lat,lng],
            tooltip=name,
            popup=v_type,
            icon= folium.features.CustomIcon(
                icon_image='https://volcano.si.edu/includes/images/volcano_icon_24.png',
                    icon_size=(20,20))
            )
     )
    ubicaciones.add_child(         # Segundo objeto: círculo interno
        folium.features.Circle(
            location=[lat,lng],
            color='red',
            fill = True,
            radius = 10000,
            popup = 'Zona de alto riesgo: 10 km'
            )
    )
    ubicaciones.add_child(         # Tercer objeto: círculo externo
        folium.features.Circle(
            location=[lat,lng],
            color='orange',
            fill = True,
            radius = 30000,
            popup = 'Zona de riesgo medio: 30 km'
            )
    )
```

Agregamos al mapa y visualizamos.

```python
mexico_map.add_child(ubicaciones)
mexico_map
```

###web scrapping
Para este mapa se requirió de técnicas de 'web scrapping' ya que la información no se encontraba disponible en el data set original y se obtuvo de [Wikipedia - Decade Volcanoes](https://en.wikipedia.org/wiki/Decade_Volcanoes).

Empezamos importando las bibliotecas requeridas para el 'web scraping'

```python
#!pip install bs4
import requests 
from bs4 import BeautifulSoup
import re
```

Obtenemos la liga, hacemos la solicitud y revisamos el estado de ésta.

```python
url = 'https://en.wikipedia.org/wiki/Decade_Volcanoes'
req = requests.get(url)
req.status_code
```

Recopilamos la información necesaria y creamos un `BS object` que nos permitirá extraer información del archivo html.

```python
data = req.text
bs_object = BeautifulSoup(data,'lxml')
```

Ubicamos la tabla dentro del objeto y creamos un DataFrame.

```python
table = bs_object.find('table')
decade_volcanos=pd.read_html(str(table), flavor='lxml')[0]
decade_volcanos
```

```python
decade_volcanos.info()
```

Vemos que este DataFrame contiene mucha información que no requerimos y sólo nos estorba, por lo que aplicaremos una limpieza de datos.

Empezamos por eliminar la última columna.

```python
decade_volcanos = decade_volcanos.drop('Affected Region/Why a Decade Volcano', axis=1)
decade_volcanos
```

Pasamos los nombres de las columnas a un formato **snake case**.
